```python
import random #Importing Random to generate or pick random card number from deck called "Cards" Below:

cards=[11,2,3,4,5,6,7,8,9,10, 10,10,10]  #Deck of random cards(numbers)


def another_card(user): #For getting another card for user or for computer from deck, randomly.
    
    card=random.choice(cards)
    
    if card==11:  #When user get an ace, then this code decide  1 or 11 based on the need of player to win..
        
        #print(f"You have got ace!")
        
        if (sum(user)+11)>21:
            
            user.append(1)
            
        else:
            user.append(11)
    else:                                     
        user.append(random.choice(cards))
    



def won_or_lost(my_cards,computer_cards):  # To calculate the score and declare who is winner at the end.
    
        
    if sum(my_cards)==sum(computer_cards)  or abs(21-sum(my_cards)) ==  abs(21-sum(computer_cards)):
        
        print(f"\nIt's a Draw!\n\nComputer cards are: {computer_cards} Computer Score is :{sum(computer_cards)}\nYour cards are: {my_cards} Your Score is :{sum(my_cards)}")

    elif sum(my_cards)==21:
              print(f'\nYou won!\n\nYour Cards are: {my_cards} Your score is :{sum(my_cards)}\nComputer Cards are: {computer_cards} Computer Score is :{sum(computer_cards)}')
       
    elif sum(computer_cards)==21:
              print(f"\nComputer won!\n\nComputer Cards are: {computer_cards} Computer score is :{sum(computer_cards)}\nYour cards are: {my_cards} Your Score is :{sum(my_cards)}")

    elif abs(21-sum(my_cards))>  abs(21-sum(computer_cards)):

        print(f"\nComputer won!\n\nComputer Cards are: {computer_cards} Computer score is :{sum(computer_cards)}\nYour cards are: {my_cards} Your Score is :{sum(my_cards)}")

    else:
        print(f'\nYou won!!\n\nYour Cards are: {my_cards} Your score is :{sum(my_cards)}\nComputer Cards are: {computer_cards} Computer Score is :{sum(computer_cards)}')
        
       

    
#starter for starting a game:
    
def start_game():
    print(logo)
    user_input=input("\nDo you want to Play Black Jack? Press 'Y' to 'YES' press 'N' for 'NO': ")
    
    if user_input.lower()=='y':
        game()
        
    else:
        while 1:
            if 1==1:
                print('\nThank you for Playing with us, See you soon!')
                break
                

#the actual game function:

def game():
    
    #Both users have empty or 0 cards in the beggining of the game..
    
    my_cards=[]
    computer_cards=[]

    #At start of the game players are getting two Cards each..

    for _ in range(2):

        another_card(my_cards)
        another_card(computer_cards)

    print(f"\nPlayer's have got two cards each:\n\nYour's Cards are: {my_cards} and your Score is: {sum(my_cards)}.\nComputer's First Card is: {computer_cards[0]}")


    #This ask user to get a card, untill user or computer have not crossed 21 or got black jack..
    
    while  ((sum(my_cards)<21) and (sum(computer_cards)<21)):

        user_input=input("\nPlease enter 'N' to get a new card, or press 'P' to pass:'")

        if user_input.lower()=='n':

            another_card(my_cards)
            print(f"\nYour cards are: {my_cards}  sum is {sum(my_cards)}\n")

            #check
            if sum(my_cards)>=21:
                break

         
        else:break
            
    #When user stop fetching the cards, then computer gets a card if its score is less then 17..

    while sum(computer_cards)<17:  
  
                print(f"\nComputer's Score is less than 17. Computer is getting another card.")
                another_card(computer_cards)
                print(f"Computer's First card is: {computer_cards[0]}")
                
            
    #After computer's score is more than 17 and user is done fetching the cards.. we calculate score and declare the winner..
    
    won_or_lost(my_cards,computer_cards)
    
    #for starting the game again: 
    
    start_game()
```
