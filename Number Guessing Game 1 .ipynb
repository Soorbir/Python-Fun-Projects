```python
import random

def set_difficulty(): #Setting difficulty level for user, 
                     #if opted for easy user get 10 attempts, if opted hard then user get 5 chances..

    Game_level= (input("\nPlease choose the difficulty level, Type 'Easy' or 'Hard': ")).lower()
    
    if Game_level=='easy':
        return 10
    elif Game_level=='hard':
        return 5
    
def check(guess,actual_number): #checking if user made the right guess or not by comparing the values..
    
    if guess==actual_number:
        print(f'\nYou have made the right Guess! the number was {actual_number}.')
        global game_end
        game_end=True

    elif guess<actual_number:
        print('You have gone too low!')

    else:print("You've gone too high!")
    
def game():  #core element of game in which set_difficulty and check functions have been used.
    
    global actual_number
    actual_number=random.randint(1,100)
    
    global game_end
    game_end=False 
    
    print(logo)
    
    print("Welcome to the Number Guessing Game!\nI'm thinking of a number between 1 and 100.")
   
    attempts=set_difficulty()

    while attempts>0:  #This loop will run untill user is left with 0 attempts
         
        print(f"\nYou have {attempts} attempts remaining to guess the number.")
        
        guess=int(input('Please enter the guess: '))
       
        check(guess,actual_number)

        attempts-=1  #as user exaust their attempt, this variable "Attempt" is going to get decremented.
        
        if game_end==True:
            break

        elif attempts==0:
            print(f'\nYou are out of chances, GAME OVER!!\nThe Actual number was {actual_number}, Better Luck next Time.')
            break
            
        else: print('Guess Again..')

    
    
    
```
